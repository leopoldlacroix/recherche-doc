```python
filtre = [",4532Z",",4520A",",4520B",",4511Z",",4519Z"]
```

```python
filepath = 'StockEtablissement_utf8.csv'
with open(filepath, encoding='utf8', errors='ignore') as fp:
   line = fp.readline()
   ctn = 1
   res = ""
   while line:
       if(ctn%10000000 == 0):
           print(ctn)
       if(any([txt in line for txt in filtre])):
           res += line
       line = fp.readline()
       ctn+=1

f = open("res.txt", "a")
f.write(res)
f.close()
print("{} lines checked".format(ctn))
       
```
